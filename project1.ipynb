# Summary of COVID-19 Clinical Trials Dataset Analysis

## Overview  
- Dataset contains 5,783 records with 27 columns, capturing various aspects of COVID-19 clinical trials globally.

## Key Findings

### Temporal Trends  
- Most trials were registered in 2020, aligning with the start of the pandemic and reflecting rapid scientific mobilization.

### Trial Phases and Scale  
- Majority of studies are in early or intermediate phases (Phase 1, Phase 2, or combined phases).  
- Enrollment sizes vary widely: many small participant groups and a few large-scale trials drive total enrollment.

### Sponsorship and Geography  
- A small number of major sponsors—hospitals, government agencies, pharmaceutical companies—account for many trials.  
- Research activity is concentrated in certain countries and institutions.

### Data Completeness and Quality  
- Core fields like trial status, study type, and sponsor information are well-populated.  
- Fields such as interventions and outcome measures have substantial missing data, limiting in-depth analysis.

## Conclusion  
- The dataset documents a globally coordinated, intense research effort against COVID-19 with concentrated sponsorship and geographic focus.  
- Future work would benefit from improved data completeness, particularly for interventions, outcomes, and detailed trial results.


```python
import pandas as pd
import numpy as np

# Load the CSV file (replace filename with the path to your downloaded dataset)
df = pd.read_csv(r"C:\Users\VAMSI\Downloads\project\project\COVID clinical trials.csv")

# Preview the structure
print(df.shape)
df.head()
```

```python
# Check data types and missing values
df.info()
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# Basic statistics for numerical columns
df.describe()
```

```python
# Visualize missing data
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,7))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# Percentage missing per column
missing_percent = df.isnull().mean().sort_values(ascending=False) * 100
print(missing_percent[missing_percent > 0])
```

```python
df.info()
```

```python
# How many unique interventions, sponsors, and phases?
print("Unique interventions:", df['Interventions'].nunique())
print("Unique sponsors:", df['Sponsor/Collaborators'].nunique())
print("Phases:", df['Phases'].unique())
print("Study Types:", df['Study Type'].unique())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='Status', data=df, order=df['Status'].value_counts().index)
plt.title('Number of Trials per Status')
plt.xlabel('Number of Trials')
plt.ylabel('Status')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(y='Phases', data=df, order=df['Phases'].value_counts().index)
plt.title('Number of Trials per Phase')
plt.xlabel('Number of Trials')
plt.ylabel('Phase')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Enrollment'].dropna(), bins=30, log=True)
plt.title('Enrollment Size Distribution (log scale)')
plt.xlabel('Enrollment Size')
plt.show()

print(df['Enrollment'].describe())
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['Primary Completion Date'] = pd.to_datetime(df['Primary Completion Date'], errors='coerce')

# Trials started per month
trials_per_month = df['Start Date'].dt.to_period('M').value_counts().sort_index()
plt.figure(figsize=(14,5))
trials_per_month.plot(kind='bar')
plt.title('Trials Registered Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Trials')
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

all_conditions = df['Conditions'].dropna().str.split(';|,').sum()
common_conditions = Counter([cond.strip() for cond in all_conditions if cond]).most_common(15)

labels, counts = zip(*common_conditions)
plt.figure(figsize=(10,7))
sns.barplot(y=list(labels), x=list(counts), orient='h')
plt.title('Most Common Conditions Studied')
plt.xlabel('Number of Trials')
plt.ylabel('Condition')
plt.tight_layout()
plt.show()
```
